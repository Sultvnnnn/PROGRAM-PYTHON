```python
import sys
import statistics as stat

gradeDict = {}

def enterGrades():
  name = input("Enter the student's Name: ")
  grade = float(input("Enter the student's Grade: "))

  if name in gradeDict:
    gradeDict[name].append(grade)
  else:
    gradeDict[name] = [grade]

  print('\nAdding Grade...')

  print("-" * 20)
  for key, value in gradeDict.items():
    print(f"{key}: {value}")
  print("-" * 20)

def removeStudent():
  name = input("Enter the student's Name: ")

  if name in gradeDict:
    print('Removing Student...')
    del gradeDict[name]
    print(f"{name} has been removed.")
  else:
    print(f"{name} was not found.")

def studentAverage():
  name = input("Enter the student's Name: ")

  if name in gradeDict:
    grades = gradeDict[name]
    average = stat.mean(grades)
    print('Calculating Average...')
    print(f"{name} has an average grade of {average}.")
  else:
    print(f"{name} was not found.")

def showNameList():
  print('Showing Name List...')
  print("-" * 20)
  for key, value in gradeDict.items():
    print(f"{key}: {value}")
  print("-" * 20)

while True:
  print('''
  Welcome to Grade Central

  [1] - Enter Grades
  [2] - Remove Student
  [3] - Student Average Grades
  [4] - Show Name List
  [5] - Exit
  ''')
  data = input("What would you like to do? (Enter a number): ")

  if data == "1":
    enterGrades()
  elif data == "2":
    removeStudent()
  elif data == "3":
    studentAverage()
  elif data == "4":
    showNameList()
  elif data == "5":
    sys.exit("Exiting Program... See You!")
  else:
    sys.exit('Invalid Input!')
```
